```python
import pandas as pd

census_tracts = pd.read_csv("csv/Census_Tracts_in_2020.csv")
census_tracts.head()
```

```python
total_pop = census_tracts["P0010001"].sum()
print("Total pop: ", total_pop)
```

```python
census_tracts["pop_percent"] = census_tracts["P0010001"] / total_pop
census_tracts.head()
```

```python
# Get list of all columns
cols = census_tracts.columns.tolist()

# Remove pop_percent from its current position
cols.remove('pop_percent')

# Find the index of GEOID and insert pop_percent after it
geoid_index = cols.index('GEOID')
cols.insert(geoid_index + 1, 'pop_percent')

# Reorder the DataFrame columns
census_tracts = census_tracts.reindex(columns=cols)
```

```python
census_tracts.head()
```

```python
census_tracts = census_tracts.rename(columns={'P0010001': 'total_pop'})
census_tracts.head()
```

```python
census_tracts.drop(["P0010002"], axis=1, inplace=True)

census_tracts = census_tracts.rename(columns={
    'P0010003': 'white',
    'P0010004': 'black',
    'P0010005': 'american_indian',
    'P0010006': 'asian',
    'P0010007': 'pacific_islander',
    'P0010008': 'other',
    'P0020002': 'hispanic'
})

census_tracts.head()
```

```python
print(list(census_tracts.columns))
```

```python
cols_to_drop = [col for col in census_tracts.columns if col.startswith('P00')]
census_tracts = census_tracts.drop(columns=cols_to_drop)
print(list(census_tracts.columns))
```

```python
census_tracts.drop(["H0010001", "H0010002", "H0010003", "POP100", "HU100"], axis=1, inplace=True)
census_tracts.head()

```

```python
print(list(census_tracts.columns))
```

```python
# Export to CSV in the public/data folder
census_tracts.to_csv('../public/data/cleaned_census_tracts.csv', index=False)
```
